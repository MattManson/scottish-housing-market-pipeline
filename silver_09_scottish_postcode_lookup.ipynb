```python
from pyspark.sql.functions import col

spark.sql("USE CATALOG scottish_housing_ws")
spark.sql("USE SCHEMA silver")
```

```python
df = spark.table("bronze.postcode_lookup")

df_silver = (
    df
    .withColumnRenamed("Postcode", "postcode")
    .withColumnRenamed("CouncilArea2019Code", "council_area_code")
    .withColumnRenamed("DataZone2022Code", "data_zone_2022")
    .withColumnRenamed("UrbanRural6Fold2022Code", "urban_rural_class")
    .withColumnRenamed("ScottishIndexOfMultipleDeprivation2020Rank", "simd_2020_rank")
)

df_silver.show(5, truncate=False)
df_silver.printSchema()
```

```python
(
    df_silver.write
    .mode("overwrite")
    .format("delta")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.postcode_lookup")
)
```
